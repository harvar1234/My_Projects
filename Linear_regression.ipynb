```python
def attribute_info(df_feature):
  print('Description features : ','\n')
  print(df_feature.describe(),'\n')
  print('Value count of attributes : ','\n')
  print(df_feature.value_counts(),'\n')
  print("Unique values of attributes : ",'\n')
  print(df_feature.unique(),'\n')
  print("Missing values of attributes : ",'\n')
  print(df_feature.isnull().sum(),'\n')
  print("% missing values of atttributes : ",'\n')
  print(df_feature.isnull().sum()/len(df_feature),'\n')
```

```python
attribute_info(data['furnishingstatus'])
```

```python
for i in data:
  print("Attribute name : ",i,'\n')
  attribute_info(data[i])
```

```python
x = data.iloc[:,1:]
y = data['price']
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
plt.scatter(y_train,y_train_pred)
rcParams['figure.figsize']=6,4
```
