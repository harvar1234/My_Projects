```python
def fetch_cast(obj):
 lst = []
 counter = 0

 for i in ast.literal_eval(obj):
  # print(i['name'])
  if counter != 3:
    lst.append(i['name'])
    counter += 1
  else:
    break

 return lst
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer#Stemming is a technique used to reduce an inflected word down to its word stem
ps = PorterStemmer()                                 #A stemming algorithm reduces the words “chocolates”, “chocolatey”, and “choco” to
                                                                  # the root word, “chocolate” and “retrieval”, “retrieved”, “retrieves” reduce to the stem “retrieve”.
```

```python
def stem(text):

 lst = []

 for i in text.split():
  lst.append(ps.stem(i))

 return " ".join(lst)
```

```python
stem('Hello buddy, I am a actions stutments and loving person addict addictive addicted')
```

```python
from sklearn.feature_extraction.text import CountVectorizer # Countvectorizer==>used to transform a given text into a vector
                                                                              # on the basis of the frequency (count) of each word that occurs in the entire text
```

```python
def get_recommended_movie(movie_name):

 movie_index = new_df[new_df['title'] == movie_name].index[0]
 distance = similarity[movie_index]
 movie_list = sorted(list(enumerate(distance)), reverse=True, key= lambda x: x[1])[1:6]
 for i in movie_list:
  print(new_df.iloc[i[0]].title)
```
